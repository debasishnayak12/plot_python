```python
# exploring all categories of each column
#categories of Gender 
print("Categories in Gender :",df["gender"].unique())

print("Categories in Review :",df["review"].unique())

print("Categories in education :",df["education"].unique())

print("Categories in purchase :",df["purchased"].unique())
```

```python
plt.figure(figsize=[10,5])
sns.histplot(data=df,x="age",color="g",kde=True)
```

most of the people from nearly 20 or nearly 70

```python
# Gender

plt.figure(figsize=[10,4])
sns.countplot(df,x="gender",palette="Set1")
```

Female are more than male

```python
f,ax=plt.subplots(1,2,figsize=[20,10])
#plt.subplot(121)
sns.countplot(df,x="review",hue="gender",ax=ax[0])
#plt.subplot(122)
sns.countplot(df,x="education",hue="gender",ax=ax[1])

for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)

for container in ax[1].containers:
    ax[1].bar_label(container,color='red',size=20)
```

```python
plt.figure(figsize=[10,5])
#plt.rcParams['figure.figsize'] = (30, 12)
plt.subplot(1, 4, 1)
size = df['gender'].value_counts()
labels = 'Female', 'Male'
color = ['red','green']

plt.pie(size, colors = color, labels = labels,autopct = '%2.f%%')
plt.title("Gender",fontsize=20)
plt.axis('off')


#pie plot for review
#plt.figure(figsize=[10,5])
plt.subplot(1, 4, 2)
size=df["review"].value_counts()
labels='Average', 'Poor', 'Good'
color="red","yellow","blue"

plt.pie(size, colors = color, labels = labels,autopct = '%2.f%%')
plt.title("Review",fontsize=20)
plt.axis('off')



#pie plot for education
#plt.figure(figsize=[10,5])
plt.subplot(1, 4, 3)
size=df["education"].value_counts()
labels='School', 'UG', 'PG'
color="yellow","green","pink"

plt.pie(size, colors = color, labels = labels,autopct = '%2.f%%')
plt.title("Education",fontsize=20)
plt.axis('off')

#pie plot for purchased
#plt.figure(figsize=[10,5])
plt.subplot(1, 4, 4)
size=df["purchased"].value_counts()
labels='Yes','No'
color="Red","blue"

plt.pie(size, colors = color, labels = labels,autopct = '%2.f%%')
plt.title("Purchased",fontsize=20)
plt.axis('off')

plt.tight_layout()
plt.grid()

plt.show()
```

From the above plot we can observe ,
- No.of female customers are more than male customers
- most of customers are poor or average ,less of them are good
- almost same no.of customers (32%) are educated from UG and PG and 36% educatd from school
- almost 50-50 product purchased and not-purchased

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['education'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['education'].value_counts(),labels=['School', 'UG', 'PG'],explode=[0.1,0,0],autopct='%1.0f%%',shadow=True,colors=['#ff4d4d','#ff8000','#7f11e0'])
plt.show()
```

```python
#countplot for review
#plt.rcParams['figure.figsize'] = (12, 6)
plt.figure(figsize=(10,6),facecolor='purple')
plt.style.use('seaborn-talk')
sns.countplot(df,x="review", palette = 'PuBu')
plt.title('Comparison of different types of review', fontweight = 30, fontsize = 20)
plt.xlabel('Types of review')
plt.ylabel('count')
plt.show()
```

```python
#plt.rcParams['figure.figsize'] = (15, 9)
plt.figure(figsize=[10,6],facecolor='yellow')
plt.style.use('fivethirtyeight')
sns.countplot(df,x='education', palette = 'Blues')
plt.title('Comparison of Education', fontweight = 30, fontsize = 30)
plt.xlabel('Education')
plt.ylabel('count')
plt.show()
```
